```python
data.drop(columns="Not_Useful_Column")
```

```python
data["Last_Name"]=data["Last_Name"].str.lstrip("...")
data["Last_Name"]=data["Last_Name"].str.lstrip("/")
data["Last_Name"]=data["Last_Name"].str.rstrip("_")
data
```

```python
data["Phone_Number"]=data["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
data
```

```python
data["Phone_Number"] = data["Phone_Number"].astype(str).str.replace('[^a-zA-Z0-9]', '', regex=True)
data
```

```python
data["Phone_Number"] = data["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]  )
data
```

```python
data["Phone_Number"] = data["Phone_Number"].str.replace('nan--','')
data["Phone_Number"] = data["Phone_Number"].str.replace('Na--','')
data
```

```python
data[["Street_Address", "State", "Zip_Code"]] = data["Address"].str.split(',', n=2, expand=True)
data
```

```python
data = data.drop("Address", axis=1)
data
```

```python
data["Paying Customer"]=data["Paying Customer"].str.replace('Yes','Y')
data["Paying Customer"]=data["Paying Customer"].str.replace('No','N')
data["Paying Customer"]=data["Paying Customer"].str.replace('Y','Yes')
data["Paying Customer"]=data["Paying Customer"].str.replace('N','No')
data
```

```python

data
```

```python
data["Do_Not_Contact"]=data["Do_Not_Contact"].str.replace('Yes','Y')
data["Do_Not_Contact"]=data["Do_Not_Contact"].str.replace('No','N')
data["Do_Not_Contact"]=data["Do_Not_Contact"].str.replace('Y','Yes')
data["Do_Not_Contact"]=data["Do_Not_Contact"].str.replace('N','No')
data
```

```python
data=data.replace('N/a','')
data=data.replace('No/a','')
data=data.replace('NaN','')
data=data.replace('None','')
data
```

```python
data["State"] = data["State"].fillna('')
data["Zip_Code"] = data["Zip_Code"].fillna('')
data
```
